**Task 1:** **Expectation** and **Variance** (a quick review)

**(1) Linear property of Expectation:** denote $X$ as the random variable, for any constants $a, b\in \mathbb{R}$, the expectation of $Y=aX+b$ is:
$$\mathbf{E}[aX+b]=a\mathbf{E}[X]+b$$

**(2) Expectation of two random variables $X_1, X_2$:** 

$$\mathbf{E}[X_1+X_2]=\mathbf{E}[X_1]+\mathbf{E}[X_2]$$

If $X_1$ $ X_2$ and $ X_3$ have the same distribution (with the same mean $\mathbf{E}(X_1)$ and variance $\text{Var}(X_1$), we can further have:

$$\mathbf{E}[X_1+X_2+X_3]=\mathbf{E}[X_1]+\mathbf{E}[X_2]+\mathbf{E}[X_3]=3\cdot \mathbf{E}[X_1]$$


**(3) Definition of Variance:** the variance of a random variable $X$ is the expectation of the random variable $Y$, where $Y=(X-\mathbf{E}[X])^2$:

$$\text{Var}(X)=\mathbf{E}\Big[(X-\mathbf{E}[X])^2\Big]$$

**(4) The relationship between $\text{Var}(X)$ and $\mathbf{E}[X]$:** $$\text{Var}(X)=\mathbf{E}[X^2]-(\mathbf{E}[X])^2$$

**(5) Property 1 of Variance:** for any constants $a, b\in \mathbb{R}$,

$$\text{Var}(aX+b)=a^2\text{Var}(X)$$

Let $a=\dfrac{1}{3}, b=0, X=X_1+X_2+X_3$, you will have:

$$\text{Var}(aX+b)=\text{Var}\Big(\dfrac{X_1+X_2+X_3}{3}\Big)=(\dfrac{1}{3})^2\text{Var}(X)=(\dfrac{1}{3})^2\text{Var}(X_1+X_2+X_3)$$

**(6) Variance of two random variables $X_1, X_2$:** $$\text{Var}[X_1+X_2]=\text{Var}[X_1]+2\cdot \text{Cov}[X_1, X_2]+\text{Var}[X_2]$$

Typically, if $X_1$ and $X_2$ are independent, 

$$\text{Cov}[X_1, X_2]=\mathbf{E}[X_1X_2]-\mathbf{E}[X_1]\mathbf{E}[X_2]=0$$

And we have: $\text{Var}[X_1+X_2]=\text{Var}[X_1]+\text{Var}[X_2]$.


If $X_1$ and $ X_2$ have the same distribution (with the same mean $\mathbf{E}(X_1)$ and variance $\text{Var}(X_1$), we can further have:  $$\text{Var}[X_1+X_2]=2\cdot \text{Var}[X_1]$$

**(7) Variance of three random variables $X_1, X_2, X_3$:**

 $$\text{Var}[X_1+X_2+X_3]= \text{Var}[X_1]+\text{Var}[X_2]+\text{Var}[X_3]+\text{Cov}[X_1, X_2]+\text{Cov}[X_1, X_3]+\text{Cov}[X_2, X_3]$$

**Task 2**. Review how to derive $E(X)$ and $\text{Var$(X)$}$ for a sample distribution: 


**Poisson Distribution**

A discrete random variable X is said to have a Poisson distribution, with parameter ${\displaystyle \lambda >0}$, if it has a probability mass function given by
$$ {f_X}(x) = \dfrac{\lambda^x e^{-\lambda}}{x!}$$


**Derivation for E(X):**


The expected value of a discrete random variable is defined as 

$$E(X)= \sum_{x \in X} x.{f_X}(x)$$

With the probability mass function of the Poisson distribution, we will have:

\begin{align}
E(X) &= \sum_{x=0}^{\infty} x.\dfrac{\lambda^x e^{-\lambda}}{x!} \\
&= \sum_{x=1}^{\infty} x.\dfrac{\lambda^x e^{-\lambda}}{x!} \\
&= e^{-\lambda}.\sum_{x=1}^{\infty} x.\dfrac{\lambda^x}{x!} \\
&= \lambda e^{-\lambda}.\sum_{x=1}^{\infty} \dfrac{\lambda^{x-1}}{(x-1)!}
\end{align}

Substituting $z=x-1$, we have $x= z+1$, and we get:

$$ E(X) = \lambda e^{-\lambda}.\sum_{z=0}^{\infty} \dfrac{\lambda^{z}}{z!} $$

using the power series expansion of the exponential function:

$$e^x = \sum_{n=0}^{\infty}\dfrac{x^n}{n!} $$


We will have:

\begin{align}
E(X) &= \lambda e^{-\lambda}.\sum_{z=0}^{\infty} \dfrac{\lambda^{z}}{z!} \\
&= \lambda e^{-\lambda}.e^\lambda \\
&= \lambda
\end{align}

**Derivation for Var(X):**

$$ \text{Var(X)}= E[X^2]-({E}[X])^2$$

We know that:
\begin{align}
E(X^2)&= \sum_{x \in X} x^2.{f_X}(x) \\
&= \sum_{x=0}^{\infty} x^2.\dfrac{\lambda^x e^{-\lambda}}{x!} \\
&= \lambda e^{-\lambda} \sum_{x=1}^{\infty} x.\dfrac{\lambda^{x-1}}{(x-1)!} \\
&= \lambda e^{-\lambda}\left(\sum_{x=1}^{\infty} (x-1).\dfrac{\lambda^{x-1}}{(x-1)!} + \sum_{x=1}^{\infty} \dfrac{\lambda^{x-1}}{(x-1)!}\right) \\
&= \lambda e^{-\lambda}\left(\lambda\sum_{x=2}^{\infty} \dfrac{\lambda^{x-2}}{(x-2)!} + \sum_{x=1}^{\infty} \dfrac{\lambda^{x-1}}{(x-1)!}\right)
\end{align}

Substituting $i= x-2$ and $j=x-1$, we will have:
\begin{align}
E(X^2)&= \lambda e^{-\lambda}\left(\lambda\sum_{i=0}^{\infty}\dfrac{\lambda^{i}}{i!}+ \sum_{j=0}^{\infty}\dfrac{\lambda^{j}}{j!} \right)\\
&= \lambda e^{-\lambda}(\lambda e^\lambda +e^\lambda) \\
&= \lambda^2 +\lambda
\end{align}

Therefore:
\begin{align}
\text{Var(X)}&= E[X^2]-({E}[X])^2 \\
&= \lambda^2 +\lambda- \lambda^2 \\
&= \lambda
\end{align}
