### **Task 4:** Using **Pandas**, **Seaborn** and **Matplotlib**

**Pandas** is a popular Python library for exploratory data analysis.

We will explore a subset of ``IMDB-movie`` dataset from Kaggle.


Load the data

````python
from google.colab import drive
drive.mount('/content/drive')
````

````python
import pandas as pd  # conventional way to import pandas
import numpy as np   
# read CSV file from the 'dataset' subdirectory using a relative path
movies=pd.read_csv('/content/drive/My Drive/CSE242/Lab_Sessions/IMDB-Movie-Data.csv')
# display the first 3 rows
movies.head(3)
````

<code>.info()</code> gives you informations of the dataframe, such as column, data type, etc.

````python
movies.info()
````

You can observe that there exists empty values in the column Metascore.

**Primary object types:**
- **DataFrame:** consists of rows and columns (like a spreadsheet)
- **Series:** a one-dimensional labeled array capable of holding any data type (integers, strings, floating point numbers, Python objects, etc.).

Suppose you are performing a regression task. 

**Features** in this dataset:

- **Title**: the title of the film;

- **Description**: one-sentence movie summary;

- **Director**: the name of the film's director;

- **Year**: the year that the film released;

- **Runtime**: the duration of the film in minutes;

- **Votes**: number of votes;

- **Metascore**: an aggregated average of critic scores. Values are between 0 and 100. Higher scores represent positive reviews.

**Potential task**: generate a rating for a movie given above features (0-10).

````python
# check the shape of the DataFrame (rows, columns)
movies.shape
````

````python
movies.describe()
````

<code>.describe()</code> summarizes main statistics of the dataframe, such as count, mean, etc.

**How to address missing values?** 

A detailed trivial solutions are available [here](https://pandas.pydata.org/pandas-docs/stable/user_guide/missing_data.html).

Typically, for int or float type numerical columns, the empty value might be directly removed or filled with several methods in [.fillna()](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.fillna.html), or through data [interpolation](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.interpolate.html#pandas.DataFrame.interpolate).

Here, we simply use the mean value to replace the empty rows of **Metascore**.

````python
mean_metascore =movies['Metascore'].mean()
print(movies['Metascore'])
movies.Metascore.fillna(mean_metascore,inplace=True)
movies.info()
````

You can see that there are no empty values in this dataframe now.

### Task 5. Visualizing with Seaborn and Matplotlib

**Matplotlib** is a comprehensive library for creating static, animated, and interactive visualizations in Python.

**Seaborn** is a Python library for statistical data visualization built on top of Matplotlib

````python
# conventional way to import seaborn
import seaborn as sns

# allow plots to appear within the notebook
%matplotlib inline
````

[Pairplot with Seaborn](https://seaborn.pydata.org/generated/seaborn.pairplot.html): 

Plot pairwise relationships in a dataset.

````python
sns.pairplot(data=movies[:200], vars=["Votes", "Runtime (Minutes)", "Metascore","Rating"], kind='reg')
````

````python
subset = movies[movies['Year'].isin([2012, 2013, 2014])]
sns.pairplot(data=subset, vars=["Votes", "Runtime (Minutes)", "Metascore","Rating"],  hue='Year')
````

[Boxplot with Seaborn](https://seaborn.pydata.org/generated/seaborn.boxplot.html)

Draw a box plot to show distributions with respect to categories.

A box plot (or box-and-whisker plot) shows the distribution of quantitative data in a way that facilitates comparisons between variables or across levels of a categorical variable. The box shows the quartiles of the dataset while the whiskers extend to show the rest of the distribution, except for points that are determined to be “outliers” using a method that is a function of the inter-quartile range.

````python
sns.boxplot(data=movies,x='Year', y='Rating')
````

[Countplot with Seaborn](https://seaborn.pydata.org/generated/seaborn.countplot.html)

Show the counts of observations in each categorical bin using bars.

````python
sns.countplot(x="Year", data=movies)
````

<code>Relplot()</code> v.s. <code>Regplot()</code> (appeared in Question 4, Assignment 1)

**Major difference:**

[Seaborn.relplot()](https://seaborn.pydata.org/generated/seaborn.relplot.html) will plot the data. Not fit the data with a linear regression model. (Appeared in Assignment 1, question 4.1)

[Seaborn.regplot()](https://seaborn.pydata.org/generated/seaborn.regplot.html) will plot the data and a linear regression model fit. (Appeared in Assignment 1, question 4.2)

**They are not same.**
